```python
import pandas as pd
import numpy as np
```

```python
def target_confidence(numberOfAnalystOpinions):
    """Define a confidence level on the target price --> basic rule for now on number of analysts"""
    
    if numberOfAnalystOpinions == np.NaN: # DOES NOT WORK ?
        confidence = "NaN"
    
    if numberOfAnalystOpinions <= 10:
        confidence = "Low"
    
    elif numberOfAnalystOpinions <= 15:
        confidence = "Medium"
    
    else:
        confidence = "High"
        
    return confidence
```

```python
def price_differential():
    
    # Load databases
    last_estimates = pd.read_csv('./databases/last_estimates.csv')
    last_prices = pd.read_csv('./databases/last_prices.csv')
    general_info = pd.read_csv('./databases/general_info.csv')
    
    # Select relevant data
    targetMedianPrice = last_estimates[['ticker', 'targetMedianPrice', 'numberOfAnalystOpinions']]
    regularMarketPrice = last_prices[['ticker', 'regularMarketPrice']]
    general_info = general_info[['ticker', 'shortName']]
    
    # Join dfs
    differential_df = pd.merge(general_info, targetMedianPrice, on='ticker')
    differential_df = pd.merge(differential_df, regularMarketPrice, on='ticker')
    
    # Compute diff
    differential_df['absolute_diff'] = differential_df['targetMedianPrice'] - differential_df['regularMarketPrice']
    differential_df['relative_diff'] = differential_df['absolute_diff'] / differential_df['regularMarketPrice']
    
    # Set a confidence level
    differential_df['confidence'] = differential_df['numberOfAnalystOpinions'].apply(lambda x: target_confidence(x))
    
    # Sort from most undervalued to most overvalued
    differential_df = differential_df.sort_values(by='relative_diff', ascending=False)
    
    return differential_df
```

```python
price_differential()
```
